```python
import numpy as np
train_size=X_train.shape[0]
class_priors={}
for outcome in np.unique(y_train):
    outcome_count=len(y_train[y_train==outcome])
    class_priors[outcome]=outcome_count/train_size
class_priors
```

```python
feature_names=list(X_train.columns) # Assign column names to 'feature_names'
likelihoods={}
for outcome in np.unique(y_train):
    outcome_count=len(y_train[y_train==outcome])
    for feature in feature_names: # Iterate through column names in 'feature_names'
        for feat_value in np.unique(X_train[feature]):
          count=0
          for i in range(len(X_train)):
            # Replace 'feat' with 'feature' to access the correct column
            if X_train[feature][i]==feat_value and y_train[i]==outcome:
              count=count+1
likelihoods[(feature,feat_value,outcome)]=(count+1)/(outcome_count+(len(feature_names))) # Use 'feature_names' for length
```
