Anomaly Detection: Credit Card Fraud Detection

Importing the Libereries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset
credit=pd.read_csv('creditcard.csv')
```

```python
# Now Handeling the missing values with Imputation method
```

```python
# Analysing the distribution of data
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in credit.columns:
    if credit[col].isnull().sum() > 0:
        # Check if the column is numeric before calculating the mean
        if pd.api.types.is_numeric_dtype(credit[col]):
            credit[col] = credit[col].fillna(credit[col].mean())
        else:
            # Handle non-numeric columns (e.g., fill with a suitable value)
            # For example, fill with the most frequent value (mode):
            credit[col] = credit[col].fillna(credit[col].mode()[0])
            # or keep null if better
            #pass

credit.isnull().sum().sum()
```

```python
# Distribution of Transactions
df=credit['Class'].value_counts(normalize=False)
print(df)
```

Under Sampling

Build a sample data set containing similar distribution of normal transaction and the fraudulent Transaction


Number of Fraudulent Transaction-->205

```python
legit_sample=legit.sample(n=205)
```

Concatenating two DataFrame

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0) # when axix is 0 the data set is added one by one. and all the 205 values will be added below the legit_sample
# and if axis =1 then the values will be added column wise
```

Splitting  the data into Features and Targets

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into training data and testing  data by using traing test split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2) # 0.2 represent 20% of test data and 80% is training data
# and if in stratify we will not assign Y so the distribution of 0 and 1 will be different in training data and testing data
# random_state means how we want to split our data
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

we have total 410 values and 328 are in training and 82 are in testing

Model Training

Logistic Regression- it is used for Binary Classification

```python
model=LogisticRegression()
```

```python
# Training the logistic regression modely with training data
model.fit(X_train,Y_train)
```

Model Evaluation


Accuracy Score

```python
# Accuracy on training  data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```
